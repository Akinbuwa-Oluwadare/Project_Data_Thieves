```python
# importing pandas and numpy

import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#exporting file 1

land = pd.read_csv('land.csv',engine="python")
land.head()
```

```python
land.columns
```

```python
#exporting file 1

animals = pd.read_csv('combined_animals_table.csv',engine="python") 
animals.head()
```

```python
#animals["Country"]=animals["Locality"].str.extract('(^(.+?),)')

animals['Country']=animals['Locality'].str.extract('^([^,]*)')

animals
```

```python
#merge

merging=land.merge(animals, left_on='Country_name', right_on='Country')

merging
```

```python
df_environmental = merging.copy()
```

```python
df_environmental.columns
```

```python
list_drop= ['2018_threatened_birds', '2018_threatened_fishes', '2018_higher_plants', 'Locality', 'Country']
```

```python
df_environmental = df_environmental.drop(axis = 1, columns = list_drop)
```

```python
df_environmental.head()
```

#### Analysis and Plotting animals data

```python
count_status = df_environmental.assign(Status_count = df_environmental['Status'] )
```

```python
yo = count_status.groupby(['Country_name', 'Status'], as_index = False).agg({ 'Status_count': 'count'})
yo.sort_values(by='Status_count', ascending = False)
```

```python
xo = count_status.groupby(['Country_name', 'prot_perc_2018'], as_index = False).agg({ 'prot_perc_2018': 'count'})
xo.sort_values(by='prot_perc_2018', ascending = False)
```

```python
count_country=df_environmental.groupby(['Country_name'], as_index = False).agg({ 'Status': 'count'})
count_country=count_country.sort_values(by='Status', ascending = False)
```

```python
count_country2=df_environmental.groupby(['Country_name'], as_index = False).agg({ 'prot_perc_2018': 'count'})
count_country2=count_country2.sort_values(by='prot_perc_2018', ascending = False)
```

```python
#plotting

count_country.plot.bar(x='Country_name', color= 'green', title = 'Number of endangered species per Country')
```

```python
#plotting

count_country2.plot.bar(x='Country_name', color= 'green', title = 'Percentage of protected land Country')
```
