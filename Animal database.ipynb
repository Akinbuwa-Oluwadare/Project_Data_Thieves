From the cell above, We can see that 90% of `TypeSpecies` is missing and there is little to what can be done to get all these missing values. So we will be dropping the column due to large missing values.

```python
Df_copy.drop("TypeSpecies", axis = 1, inplace = True)
```

```python
Df_copy.head()
```

```python
Df_copy.isna().sum()/len(Df_copy)
```

```python
#Renaming the columns to a more suitable names so as to facilitate proper selection of the columns.

Df_copy.rename(columns = {'CommonName': 'Common_name',
                         'TypeLocality': 'Locality',
                         'Extinct?': 'Extinct'}, 
                      inplace = True)
```

```python
Df_copy.head()
```

Investigating each columns so as to know how to deal with the missing values.

```python
Df_copy["Species"].value_counts()
```

```python
Df_copy["Species"].count()
```

```python
Df_copy.loc[Df_copy.Species.isnull(), ["Family","Species", "Common_name"]]
```

```python
# Since it is difficult to get the Species of the animals from the dataframe while comparing it with other columns
# So we will be dropping the rows with the missing values.

Df_copy.dropna(subset = ["Species", "Status"],how = "any",inplace = True)
```

Handling other Columns with null values

```python
Df_copy.isna().sum()
```

```python
# looking into Locality and investigating the missing values.

Df_copy.loc[Df_copy.Locality.isnull(),"Common_name"]
```

```python
# I will be dropping the locality with null values because i am not sure of the animals locality.
Df_copy.dropna(subset = ["Locality", "Common_name"],how = "any",inplace = True)
```

```python
Df_copy.reset_index(drop = True).head()
```

```python
Df_copy.shape
```

```python
#Getting ride of the encode in the dataframe.
def remove_non_ascii(text):
    return ''.join([i for i in str (text) if ord(i)<128])
Df_copy["Status"] = Df_copy["Status"].apply(remove_non_ascii)
```

```python
Df_copy["Locality"] = Df_copy["Locality"].apply(remove_non_ascii)
```

```python
Df_copy["Common_name"] = Df_copy["Common_name"].apply(remove_non_ascii)
```

```python
Df_copy.reset_index(drop = True).head(50)
```
